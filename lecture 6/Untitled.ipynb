```python
import matplotlib.pyplot as plt
import cv2 # нужно поставить себе cv2 через conda install opencv
import numpy as np

def draw_picture(image, bgr=False):
    b, g, r = cv2.split(image) # по умолчанию cv2 почему-то отдает цвета в порядке BGR вместо RGB
    new_image = cv2.merge([r, g, b])
    plt.figure(figsize=(7, 5))
    plt.axis('off')
    plt.imshow(new_image)
    plt.show()
```

```python
img = cv2.imread('birds.jpg') # считать картинку
#draw_picture(img) # нарисовать картинку
```

```python
def draw_pictures(images, titel):
    plt.figure(figsize=(15, 10))
    
    for i in range(len(images)) :
        image = images[i]
        b, g, r = cv2.split(image)
        new_image = cv2.merge([r, g, b])
        plt.subplot(220 + i + 1)
        plt.axis('off')
        plt.imshow(new_image)
        plt.title(str(i + 1))
    
    plt.show()
```

```python
images = np.array([img, img, img, img])
draw_pictures(images)
```

```python
snake_img = cv2.imread('snake.jpg')
draw_picture(snake_img)
```
